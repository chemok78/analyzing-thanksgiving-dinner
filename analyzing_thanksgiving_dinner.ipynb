```python
import pandas as pd
data = pd.read_csv("thanksgiving.csv", encoding="Latin-1")
print(data.head(5))
```

```python
# display value counts of celebrate yes/no
print(data["Do you celebrate Thanksgiving?"].value_counts())
```

```python
# Boolean array where celebrate is Yes
celebrate = data["Do you celebrate Thanksgiving?"] == "Yes"
# filter out where celebrate is No
data_dropped_no = data[celebrate]
print(data_dropped_no["Do you celebrate Thanksgiving?"].value_counts())
```

```python
# create Boolean lists where value in column is null or not
apple_isnull = data["Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Apple"].isnull()
pumpkin_isnull = data["Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pumpkin"].isnull()
pecan_isnull = data["Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pecan"].isnull()
# combine 3 isnull values for every row
# if one is True, then it returns True. Otherwise False
# ate_pies returns a list of people who didnt eat pie
ate_pies = data[apple_isnull & pumpkin_isnull & pecan_isnull]
print(ate_pies)
```

```python
def convert(string):
    if pd.isnull(string) == True:
        return None
    else:
        split = string.split()
        first = split[0]
        first = first.replace("+", "")
        return int(first)
    
data["int_age"] = data["Age"].apply(convert)
#print(data["int_age"].head(5))
#print(int_age)
print(data["int_age"].describe())
#average age of data set is 39.38 years
#all respondents are 18+ and oldest is 60 or plus
#we counted 1025 people
    
    
```

```python
print(data.columns)
```

```python
def convert_income(string):
    if pd.isnull(string) == True:
        return None
    else:
        split = string.split()
        income = split[0]
        if income == "Prefer":
            return None
        else:
            income = income.replace("$", "")
            income = income.replace(",", "")
            return int(income)
data["int_income"] = data["How much total combined money did all members of your HOUSEHOLD earn last year?"].apply(convert_income)
print(data["int_income"].describe())
```

The mean income of 889 respondents is 74077 USD

```python
travel_distance= data["How far will you travel for Thanksgiving?"]
print("Low income and travel distance:")
low_income = data["int_income"] < 150000
low_income_travel = travel_distance[low_income]
print(low_income_travel.value_counts())
print(low_income_travel.describe())
print("High income and travel distance:")
high_income = data["int_income"] > 150000
high_income_travel = travel_distance[high_income]
print(high_income_travel.describe())
print(high_income_travel.value_counts())
```

The percentage of high income people staying at home is actually higher than low income people
